```python
#Abstract Syntax Trees
import ast
#Hàm convert này dùng để lấy tất cả giá trị từ khóa name trong từng phần tử của chuỗi đầu vào obj.
#ast.literal_eval(obj): Chuyển chuỗi obj thành một đối tượng Python
def convert_2251010060_NguyenThanhLong(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name']) 
    return L
```

```python
movies_2251010060_NguyenThanhLong['genres']=movies_2251010060_NguyenThanhLong['genres'].apply(convert_2251010060_NguyenThanhLong)
movies_2251010060_NguyenThanhLong['keywords']=movies_2251010060_NguyenThanhLong['keywords'].apply(convert_2251010060_NguyenThanhLong)
movies_2251010060_NguyenThanhLong.head()
```

```python
#Hàm này sẽ trả về danh sách chứa tối đa 3 tên từ khóa name trong đối tượng JSON
def convert2_2251010060_NguyenThanhLong(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name']) 
            counter += 1
        else:
            break
    return L
```

```python
movies_2251010060_NguyenThanhLong['cast']=movies_2251010060_NguyenThanhLong['cast'].apply(convert2_2251010060_NguyenThanhLong)
movies_2251010060_NguyenThanhLong.head()
```

```python
def fetch_director_2251010060_NguyenThanhLong(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])  
    return L
```

```python
movies_2251010060_NguyenThanhLong['crew']=movies_2251010060_NguyenThanhLong['crew'].apply(fetch_director_2251010060_NguyenThanhLong)
movies_2251010060_NguyenThanhLong.head()
```

```python
#thư viện Natural Language Toolkit (NLTK)
import nltk
#thuật toán cắt tỉa từ (stemming) để rút gọn từ về gốc của nó.
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_2251010060_NguyenThanhLong(text):
    y = []  # Khởi tạo danh sách rỗng để chứa các từ đã cắt tỉa
    for i in text.split():  # Tách văn bản thành từng từ
        y.append(ps.stem(i))  # Cắt tỉa mỗi từ và thêm vào danh sách y
    return " ".join(y)  # Nối các từ đã cắt tỉa thành một chuỗi
```

```python
# Hàm gợi ý phim
def recommend_2251010060_NguyenThanhLong(movie):
    # Lấy chỉ số của bộ phim trong DataFrame
    try:
        movie_index = new_movies_2251010060_NguyenThanhLong[new_movies_2251010060_NguyenThanhLong['title'] == movie].index[0]
    except IndexError:
        return ["Movie not found"]

    # Lấy danh sách các độ tương đồng với bộ phim đã cho
    distances = similarity[movie_index]

    # Sắp xếp danh sách phim theo độ tương đồng, lấy 5 phim tương tự nhất (trừ bản thân phim gốc)
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    # Tạo danh sách để lưu tên các bộ phim gợi ý
    recommended_movies = []
    for i in movies_list:
        recommended_movies.append(new_movies_2251010060_NguyenThanhLong.iloc[i[0]].title)

    return recommended_movies
```

COLLABORATIVE FILTERING

```python
# Tạo ma trận người dùng(pivot table)
ratings_pivot = ratings_2251010060_NguyenThanhLong.pivot_table(index='user_id', columns='movie_id', values='rating')
ratings_pivot
```

```python
ratings_pivot = ratings_pivot.fillna(0)
ratings_pivot 
```

```python
print(ratings_pivot.shape)
```

```python
# Tính toán ma trận tương đồng cosine giữa các bộ phim
similarity_matrix = cosine_similarity(ratings_pivot.T)  # Chuyển vị ma trận để tính similarity theo cột (movie_id)
```

```python
similarity_matrix
```

```python
# Chuyển đổi ma trận tương đồng thành DataFrame
similarity_df = pd_2251010060_NguyenThanhLong.DataFrame(similarity_matrix, index=ratings_pivot.columns, columns=ratings_pivot.columns)
```

```python
similarity_df
```

```python
def recommend_collaborative_2251010060_NguyenThanhLong(movie_name, num_recommendations=5):
    # Tìm ID của bộ phim từ tên bộ phim trong bảng ratings_2251010060_NguyenThanhLong
    movie_id = ratings_2251010060_NguyenThanhLong[ratings_2251010060_NguyenThanhLong['title'] == movie_name]['movie_id'].values

    if len(movie_id) == 0:
        return "Bộ phim không tồn tại trong cơ sở dữ liệu"

    movie_id = movie_id[0]  # Lấy ID của bộ phim (nếu có)

    # Kiểm tra xem ID bộ phim có trong ma trận tương đồng không
    if movie_id not in similarity_df.columns:
        return "Bộ phim không có trong ma trận tương đồng"

    # Lấy độ tương đồng của bộ phim với các bộ phim khác từ similarity_df
    movie_similarity = similarity_df[movie_id].sort_values(ascending=False)[1:num_recommendations+1]

    # Lấy danh sách các bộ phim tương tự (dựa trên ID bộ phim)
    recommended_movie_ids = movie_similarity.index.tolist()

    # Lấy tên bộ phim từ bảng ratings_2251010060_NguyenThanhLong
    recommended_movie_titles = ratings_2251010060_NguyenThanhLong[ratings_2251010060_NguyenThanhLong['movie_id'].isin(recommended_movie_ids)]['title'].unique().tolist()

    return recommended_movie_titles
```

```python
recommended_movies = recommend_collaborative_2251010060_NguyenThanhLong("Minions", num_recommendations=5)
print(recommended_movies)
```
